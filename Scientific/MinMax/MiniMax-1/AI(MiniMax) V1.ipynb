        <p></p>
        خب حالا باید رباتی بسازیم که بتواند بهترین بازی را ارائه بدهد! فعلا یک کلاس MinimaxBot می‌سازیم که قراره رباتمان باشد.
        <p></p>
        کلیت این کلاس آن است که داخلش یک دیکشنری داریم، این دیکشنری خودش شامل دو دیکشنری‌ست و کلا با توجه به بازیکن و وضعیت فعلی بازی بهترین حرکت را به ما معرفی می‌کند
        <p></p>
        در تابع اصلی‌مان باید الگوریتم Minimax را پیاه‌سازی کنیم و خروجی این تابع باید بهترین حرکت و عدد حالت نهایی باشد (به عبارتی score) به کامنت‌های کد دقت کنید:
    </font>
</div>

```python
from TicTacToefile import TicTacToe
import math
class MinimaxBot:
    def __init__(self):
        self.bestmove = {}

    def minimax(self, game, player):
        
        if player not in self.bestmove: 
            self.bestmove[player] = {}
        
        moves = game.get_moves()
        
        if game.gameover():
            best_move = None
            best_score = game.score_game()
        elif player == "X":
            best_score = -math.inf
            for move in moves:
                clone = game.copy()
                clone.move("X", move)
                score = self.minimax(clone, "O")[1]
                if score >= best_score:
                    best_score = score
                    best_move = move
        else:
            best_score = math.inf
            for move in moves:
                clone = game.copy()
                clone.move("O", move)
                score = self.minimax(clone, "X")[1]
                if score <= best_score:
                    best_score = score
                    best_move = move
        self.bestmove[player][game] = (best_move, best_score)
        return self.bestmove[player][game]
```

        <p></p>
        حالا که رباتمان را ساختیم؛ باید یک بازی بسازیم که بتوانیم با رباتمان بازی کنیم:

```python
game = TicTacToe()
bot = MinimaxBot()

human = None
while human is None:
    human = input("Select player: {} ".format(game.legal_players)).upper()
    if human not in game.legal_players:
        print("Invalid option")
        human = None

comp = [alt_player for alt_player in game.legal_players if alt_player != human][0]
turn = "X"
while not game.gameover():    
    if comp == turn:
        move = bot.minimax(game, comp)[0]
    else:
        print(game)
        move = None
        while move is None:
            move = input("Input move: ")
            if move.upper()[:1] == "H":
                print(game.__doc__)
                continue
            if move.isdigit():
                move = int(move)
    if move in game.get_moves():
        game.move(turn, move)
        turn = comp if turn == human else human
    else:
        print("Illegal move. H for help")
print(game)
if game.winner() == human:
    print("You won!")
elif game.winner() == comp:
    print("You lost :-(")
else:
    print("It's a draw")
```

        <p></p>
        در نتیجه ما در کلاس بازی کد زیر را به عنوان مقدار در وضعیت خاصی از بازی قرار می‌دهیم. (مثل بازی دوز برای بهتر فهمیدن روند مدل سازی بازی در کد خوب است فایل ConnectFourfile.py را مطالعه کنید.)
        <br>
       دقت کنید که بازی یک آرایه دو بعدی از $0$ و $1$ است که نمایانگر دو نوع مهره بازی اصلی هستند. $0$ نمایانگر مهره‌های نفر اول و $1$ نمایانگر مهره‌های نفر دوم است. 
      </font>

```python
def score_game(self, board):
        heur = 0
        state = self.board
        for i in range(0, self.WIDTH):
            for j in range(0, self.HEIGHT):
                # check horizontal streaks
                try:
                    # add player one streak scores to heur
                    if state[i][j] == state[i + 1][j] == 0:
                        heur += 10
                    if state[i][j] == state[i + 1][j] == state[i + 2][j] == 0:
                        heur += 100
                    if state[i][j] == state[i+1][j] == state[i+2][j] == state[i+3][j] == 0:
                        heur += 10000

                    # subtract player two streak score to heur
                    if state[i][j] == state[i + 1][j] == 1:
                        heur -= 10
                    if state[i][j] == state[i + 1][j] == state[i + 2][j] == 1:
                        heur -= 100
                    if state[i][j] == state[i+1][j] == state[i+2][j] == state[i+3][j] == 1:
                        heur -= 10000
                except IndexError:
                    pass

                # check vertical streaks
                try:
                    # add player one vertical streaks to heur
                    if state[i][j] == state[i][j + 1] == 0:
                        heur += 10
                    if state[i][j] == state[i][j + 1] == state[i][j + 2] == 0:
                        heur += 100
                    if state[i][j] == state[i][j+1] == state[i][j+2] == state[i][j+3] == 0:
                        heur += 10000

                    # subtract player two streaks from heur
                    if state[i][j] == state[i][j + 1] == 1:
                        heur -= 10
                    if state[i][j] == state[i][j + 1] == state[i][j + 2] == 1:
                        heur -= 100
                    if state[i][j] == state[i][j+1] == state[i][j+2] == state[i][j+3] == 1:
                        heur -= 10000
                except IndexError:
                    pass

                # check positive diagonal streaks
                try:
                    # add player one streaks to heur
                    if not j + 3 > self.HEIGHT and state[i][j] == state[i + 1][j + 1] == 0:
                        heur += 100
                    if not j + 3 > self.HEIGHT and state[i][j] == state[i + 1][j + 1] == state[i + 2][j + 2] == 0:
                        heur += 100
                    if not j + 3 > self.HEIGHT and state[i][j] == state[i+1][j + 1] == state[i+2][j + 2] \
                            == state[i+3][j + 3] == 0:
                        heur += 10000

                    # add player two streaks to heur
                    if not j + 3 > self.HEIGHT and state[i][j] == state[i + 1][j + 1] == 1:
                        heur -= 100
                    if not j + 3 > self.HEIGHT and state[i][j] == state[i + 1][j + 1] == state[i + 2][j + 2] == 1:
                        heur -= 100
                    if not j + 3 > self.HEIGHT and state[i][j] == state[i+1][j + 1] == state[i+2][j + 2] \
                            == state[i+3][j + 3] == 1:
                        heur -= 10000
                except IndexError:
                    pass

                # check negative diagonal streaks
                try:
                    # add  player one streaks
                    if not j - 3 < 0 and state[i][j] == state[i+1][j - 1] == 0:
                        heur += 10
                    if not j - 3 < 0 and state[i][j] == state[i+1][j - 1] == state[i+2][j - 2] == 0:
                        heur += 100
                    if not j - 3 < 0 and state[i][j] == state[i+1][j - 1] == state[i+2][j - 2] \
                            == state[i+3][j - 3] == 0:
                        heur += 10000

                    # subtract player two streaks
                    if not j - 3 < 0 and state[i][j] == state[i+1][j - 1] == 1:
                        heur -= 10
                    if not j - 3 < 0 and state[i][j] == state[i+1][j - 1] == state[i+2][j - 2] == 1:
                        heur -= 100
                    if not j - 3 < 0 and state[i][j] == state[i+1][j - 1] == state[i+2][j - 2] \
                            == state[i+3][j - 3] == 1:
                        heur -= 10000
                except IndexError:
                    pass
        return heur
```

```python
from ConnectFourfile import ConnectFour
import math

class PlayerMM:
    def __init__(self):
        self.bestmove = {}

    def minimax(self, game, player, depth):
        if player not in self.bestmove:
            self.bestmove[player] = {}
        moves = game.get_moves()
        if game.isTerminal()>-1:
            best_move = None
            if game.isTerminal() == 0:
                best_score = 0
            else:
                if game.isTerminal() == 1:
                    best_score = math.inf
                else:
                    best_score = -math.inf
        elif depth != 0:
            if player == 0:
                best_score = -math.inf
                for move in moves:
                    clone = game.copy()
                    clone.move(move, 0)
                    score =  self.minimax(clone, 1, depth-1)[1]
                    if score >= best_score:
                        best_score = score
                        best_move = move
            else:
                best_score = math.inf
                for move in moves:
                    clone = game.copy()
                    clone.move(move, 1)
                    score =  self.minimax(clone, 0, depth-1)[1]
                    if score <= best_score:
                        best_score = score
                        best_move = move
        else:  #depth == 0
            best_move = None
            best_score = game.score_game()
        self.bestmove[player][game] = (best_move, best_score)
        return self.bestmove[player][game] 
```

        <p></p>
        و دقت کنید که در ابتدای راه حل هم باید، آلفا و بتای راس مورد نظرمان را به ترتیب $-\infty$ و $\infty$ در نظر بگیریم، زیرا نه فرد ماکزیمم کننده و نه فرد مینیمم کننده نمی‌توانند چیزی را تضمین کنند! در نتیجه از بدترین حالت ممکن شروع می‌کنیم.

```python
class PlayerAB():

    def __init__(self):
        self.bestmove = {}

    def alphabeta(self, game, player, depth, alpha, beta):
        if player not in self.bestmove:
            self.bestmove[player] = {}
        if game.isTerminal() > -1:
            best_move = None
            if game.isTerminal() == 0:
                best_score = 0
            else:
                if game.isTerminal() == 1:
                    best_score = math.inf
                else:
                    best_score = -math.inf
        elif depth!=0:
            moves = game.get_moves()
            if(player==0):
                best_score = -math.inf
                best_move = None
                for move in moves:
                    clone = game.copy()
                    clone.move(move, 0)
                    score =  self.alphabeta(clone, 1, depth-1, alpha, beta)[1]
                    if score > best_score:
                        best_score = score
                        best_move = move
                    alpha = max(alpha, best_score)
                    if alpha>beta or alpha == math.inf or beta == -math.inf:
                        break
                if best_move == None:
                    best_move = moves[0]
            else:
                best_score = math.inf
                best_move = None
                for move in moves:
                    clone = game.copy()
                    clone.move(move, 1)
                    score =  self.alphabeta(clone, 0, depth-1, alpha, beta)[1]
                    if score < best_score:
                        best_score = score
                        best_move = move
                    beta = min(beta, best_score)
                    if alpha>beta or alpha == math.inf or beta == -math.inf:
                        break
                if best_move == None:
                    best_move = moves[0]
        else:
            best_move = None
            best_score = game.score_game()
        self.bestmove[player][game] = (best_move, best_score)
        return self.bestmove[player][game]
```
